2. Explorad el nombre de las columnas, ¿hay alguna que este repetida o que no nos aporte información de valor? ¿Hay alguna columna que solo tenga un valor? En caso de ser así, eliminadlas. **NOTA** Hay tres columnas que cuando exploramos sus valores únicos y sus frecuencias ayer, nos deberíamos haber dado cuenta que solo tiene un valor.

```python
#Exploramos las columnas:
df_unido.info()
```

```python
#nombres columnas:
lista_columnas=list(df_unido.columns)
print(f' El numero de columnas es : {len(lista_columnas)}')
print(f'El numero de valores unicos que hay en esa lista de columnas es:{len(set(lista_columnas))}')#usamos el set para asegurarnos de que no hay valores repetidos en las columnas. Los sets no pueden tener valores iguales.
print(lista_columnas)
```

```python
#para considerar la utilidad de las columnas sacamos sus valores unicos y frecuencia:
for columna in lista_columnas:
    print(df_unido.value_counts(columna),'\n')
```

```python
#hay tres columnas que solo tienen un tipo de valor:
lista_columnas_eliminar=[]
for columna in lista_columnas:
        if len(list (df_unido[columna].unique()))==1: # creamos un ciclo for donde indicamos que queremos las columnas que presenten un solo valor unico.
            print(df_unido.value_counts(columna),'\n')
            lista_columnas_eliminar.append(columna)
        
```

```python
#borrarlos:
df_unido_borrado=df_unido.drop(lista_columnas_eliminar,axis=1,inplace=False)
df_unido_borrado.info()
```

3. Algunas de las columnas tienen `.` en sus nombres y en otras tenemos `_`.  Unifica el nombre de las columnas para que todas tengan "_". Además, todas deben estar en minúsculas. 

```python
#creamos un dic comprehension para iterar con los nombres de la columna y crear un diccionario
#con los nombres antiguos:nuevos
nuevas_columnas={columna:columna.lower().replace('.','_') for columna in df_unido_borrado}
```

```python
df_limpio_columnas=df_unido_borrado.rename(columns=nuevas_columnas, inplace=False)#renombramos las columnas
df_limpio_columnas.head(1)
```
