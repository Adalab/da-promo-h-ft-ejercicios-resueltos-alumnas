```python
import pandas as pd # importamos libreria pandas 
```

```python
#rea una serie con nuemeros correlativos entre 1 y 12 inclusive
lista=list(range(1,13))
serie=pd.Series(lista)
print(serie)
```

```python
#Cambia el indice de la serie anteriormente creada para que comience con el numero 20.
indices=list(range(20,(20+12))) #creamos un una lista de numeros corelativos que empiece por 20 hasta 12 digitos más.
print(indices)
```

```python

serie_1=pd.Series(lista)
serie1_con_ind=pd.Series(lista, index=indices) # creamos una serie a partir de una lista, e indicamos los indices
print(serie1_con_ind)
```

```python
#Nuevamente cambia el indice de la serie, esta vez con una lista que haga referencia a los meses del año (tenemos 12 elementos como la cantidad de meses del año).
indices_año=['enero','febrero','marzo','abril','mayo','junio','julio','agosto','septiembre','octubre','noviembre','diciembre'] 
series_con_mes=pd.Series(lista,index=indices_año)
print(series_con_mes)
```
