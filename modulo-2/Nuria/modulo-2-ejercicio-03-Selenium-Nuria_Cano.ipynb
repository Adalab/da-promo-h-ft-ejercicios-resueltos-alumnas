```python
#Importamos librerias
from time import sleep
import pandas as pd
import requests
from bs4 import BeautifulSoup

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
```

```python
#Iniciamos el navegador y vamos a la página de Decathlon
driver = webdriver.Chrome()
driver.get('https://www.decathlon.es/')

#maximizamos la pantalla
driver.maximize_window()

#aceptamos las cookies
sleep(2)
driver.find_element('css selector', '#didomi-notice-agree-button > span').click()

#Preguntamos al usuario por la busqueda
busqueda = input('Que producto deseas buscar?')

#introducimos el input del usuario en la busqueda de la web y le damos al ENTER
driver.implicitly_wait(10)
driver.find_element('css selector','#search-bar > div > form > input').send_keys(busqueda, Keys.ENTER)

#creamos una lista para almacenar las url de las 5 primeras busquedas
url_busqueda = []

#itermaos para conseguir las 5 primeras url
for i in range(1,6):
    url_busqueda.append(driver.find_element('xpath', f'/html/body/div[2]/main/div[2]/section[2]/div/div[{i}]/div[3]/div[1]/div/a').get_attribute('href')) 

#cerrar el navegador
driver.close()
```

```python
#iniciamos el d¡iccionario donde guardaremos nuestros datos
diccionario_resultados = {'nombre': [], 
                          'precio': [], 
                          'puntuacion': []}

#Iteramos en la lista de las urls y añadimos los headers
headers = {'User-Agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Mobile Safari/537.36'}

for url in url_busqueda:

    #hacemos llamada a la url
    result_busqueda = requests.get(url,headers=headers)
    print(result_busqueda.status_code)

    soup_busqueda = BeautifulSoup(result_busqueda.content, 'html.parser')

    #Utilizamos un try/except ya que sabemos que hay busquedas que nos dan error
    try:
        diccionario_resultados['nombre'].append(soup_busqueda.find('h1').text)
    except:
        diccionario_resultados['nombre'].append('error')

    try:
        diccionario_resultados['precio'].append(''.join(soup_busqueda.find('span', {'class':'vtmn-price vtmn-price_size--large vtmn-price_variant--accent'}).text.split()))
    except:
        diccionario_resultados['precio'].append('error')

    try:    
        diccionario_resultados['puntuacion'].append(soup_busqueda.find('span', {'class':'vtmn-rating_comment--primary'}))
    except:     
        diccionario_resultados['puntuacion'].append('error')

    
```

```python
#creamos nuestro DataFrame a partir del diccionario de resultados
df_busqueda = pd.DataFrame(diccionario_resultados, index=None)
df_busqueda
```
