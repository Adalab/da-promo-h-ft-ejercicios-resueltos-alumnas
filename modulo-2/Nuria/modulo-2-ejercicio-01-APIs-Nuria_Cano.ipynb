- Create the dictionary where we will store the data for all players

```python
dic_info_players = {'nombres':[], 'apellidos': [], 'equipo':[]}
```

- Create a function to organize the obtained data.

```python
def limpiar_info_player (datos_json, dic_resultados):

    '''
    Esta función toma datos en formato JSON de un jugador y los organiza en un diccionario de resultados.

    Parameters:
    ----------
    - datos_json(dict):Los datos JSON de los jugadores que se van a procesar.
    - diccionario_resultados (dict): Un diccionario que almacena los resultado.

    Returns:
    --------
    - diccionario_resultados (dict): Un diccionario actualizado que contiene información sobre el jugador.  
    '''

    for player in datos_json['data']:
        dic_resultados['nombres'].append(player['first_name'])
        dic_resultados['apellidos'].append(player['last_name'])
        dic_resultados['equipo'].append(player['team']['full_name'])

    return dic_resultados


```

```python
dic_info_players = limpiar_info_player(json_all_players, dic_info_players)
dic_info_players
```

- convert the organized data into DataFrame.

```python
df_info_all_players = pd.DataFrame(dic_info_players)
df_info_all_players
```

* Iterate in the list to acces the info of each dictionary:

```python
for team in js_teams['data']:
    print(team['id'])
    print(team['conference'])
    print(team['division'])
    print(team['city'])
    print(team['name'])
    print(team['full_name'])
    print(team['abbreviation'])
    break
```

```python
dict_teams_info = {'id': [],
                   'conference': [],
                   'division': [],
                   'city': [],
                   'name': [],
                   'full_name': [],
                   'abbreviation': []}
```

```python
def clean_info_teams (json_data, dict_results):

    '''
    This function organizes into a dictionary JSON info about basketball teams.

    Parameters:
    -----------
    - json_data (dict): The JSON data about all teams in.
    - dict_results (dict): A dictionary that will store the sorted data.

    Return:
    ------
    - dict_results (dict): The updated dictionary with all the info 
    '''

    for team in json_data['data']:
        dict_results['id'].append(team['id'])
        dict_results['conference'].append(team['conference'])
        dict_results['division'].append(team['division'])
        dict_results['city'].append(team['city'])
        dict_results['name'].append(team['name'])
        dict_results['full_name'].append(team['full_name'])
        dict_results['abbreviation'].append(team['abbreviation'])

    return dict_results
```

```python
dict_teams_info = clean_info_teams(js_teams, dict_teams_info)
dict_teams_info
```

```python
df_teams_info = pd.DataFrame(dict_teams_info)
df_teams_info[:10]
```
